<span style=color:green>Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.</span>

Ans-

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.rand(3)
x = 4+np.random.normal(0,2,50)
y = 3+np.random.normal(0,2, len(x))

plt.scatter(x,y)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter plot")
plt.show()
```

<span style=color:green>Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:</span>

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line Plot 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line Plot 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line Plot 4')


plt.tight_layout()


plt.show()
```

<span style=color:green>Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])</span>

```python
import numpy as np
import matplotlib.pyplot as plt

company=np.array(["Apple","Microsoft","Google","AMD"])
profit=np.array([3000,8000,1000,10000])

# horizontal bar plot
plt.barh(company,profit)
plt.xlabel('profit')
plt.ylabel('company')
plt.title('horizontal')
plt.show()

# bar plot
plt.bar(company,profit)
plt.xlabel('profit')
plt.ylabel('comapny')
plt.title('bar plot')
plt.show()
```
